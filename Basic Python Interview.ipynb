#### Q1. Write a code to convert a string into list
##### Example: 
Input:'Hello World' -> Output: ['Hello','World']

#### Answer

```python
#function
def string_to_list(strn):
    return strn.split(' ')
```

```python
#test the function by yourself
strn=input('enter the phrase')
print(string_to_list(strn))
```

#### Q2. Write a code to reverse a string
##### Example:
Input:'Hello World' -> Output: 'dlroW olleH'

#### Answer 1

```python
#function
def reverse_string_1(strn):
    return strn[::-1]
```

```python
# test the function by yourself
strn=input('Enter the String: \n')
print(reverse_string_1(strn))
```

#### Answer 2

```python
# function
def reverse_string_2(strn):
    rev_strn=''
    for s in strn:
        rev_strn=s+rev_strn
    return rev_strn
```

```python
# test the function by yourself
strn=input('Enter the String: \n')
print(reverse_string_2(strn))
```

#### Q3. Order a list of numbers in ascending or descending mode
##### Example:
Input: [1,5,2,4,1],ascending -> Output: [1,1,2,4,5]

#### Answer

```python
#function
def sort_list(lst,asc_desc='asc'):
    if asc_desc=='desc':
        return sorted(lst,reverse=True)
    return sorted(lst)
```

```python
lst = [1,5,2,4,1]
print(sort_list(lst))
print(sort_list(lst,asc_desc='desc'))
```

#### Q4. Find the minimum value in a list of numbers and remove it from the list. If the number's shown up multiple times, remove all of them

##### Example:
Input: [1,5,2,4] -> Output: [5,2,4] \
Input: [1,5,2,4,1] -> Output: [5,2,4]

#### Answer

There could be a couple of solutions, but here we come up with two: \


```python
#first solution
def remove_minimums(lst):
    new_list=[]
    m=min(lst)
    for num in lst:
        if num!=m:
            new_list.append(num)
    return new_list          
```

```python
remove_minimums([1,2,3,4,1,2,1,3,6,-4])
```

#### Q5.  Write a code in pythontakes a string as input and returns a list of characters that are repeated multiple times in the string using basic functions.

#### Answer

```python
def find_repeated_characters(input_string):
    # Create an empty dictionary to count character occurrences
    char_count = {}
    
    # Iterate over each character in the string
    for char in input_string:
        # Update the count of each character in the dictionary
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    
    # Create a list of characters that are repeated more than once
    repeated_chars = [char for char, count in char_count.items() if count > 1]
    
    return repeated_chars

# Example usage
input_string = "discombobulated"
repeated_characters = find_repeated_characters(input_string)
print("Repeated characters:", repeated_characters)
```
